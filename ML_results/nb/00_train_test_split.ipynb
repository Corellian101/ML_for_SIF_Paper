# Imports

```python
import numpy as np
import sklearn
import pandas as pd

from sklearn.model_selection import train_test_split
```

# Dataset 1: RN Dataset

```python
data = pd.read_csv("../data/3_datasets_new/1_RN_data.csv")
data
```

```python
train, test = train_test_split(data, test_size=0.25, shuffle=True)
train.shape, test.shape
```

```python
train
```

```python
test
```

```python
train.to_csv('../data/3_datasets_new/1_RN_data_TRAIN.csv', encoding='utf-8', index=False)
test.to_csv('../data/3_datasets_new/1_RN_data_TEST.csv', encoding='utf-8', index=False)

train.to_csv('../data/3_datasets_new/1_RN_data_TRAIN_idx.csv', encoding='utf-8', index=True)
test.to_csv('../data/3_datasets_new/1_RN_data_TEST_idx.csv', encoding='utf-8', index=True)
```
